```python
# Step 1: Get total sales per city and sort in descending order
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)

# Step 2: Pick top 10 cities
top_10_cities = city_sales.head(10).index.tolist()

# Step 3: Calculate sub-category sales for those cities
df_top10 = df[df['City'].isin(top_10_cities)]

# Step 4: Iterate over top cities in order and print their top 5 sub-categories
for city in top_10_cities:
    total_sales = city_sales[city]

    # Filter data for this city and get top 5 sub-categories
    city_data = (
        df_top10[df_top10['City'] == city]
        .groupby('Sub-Category')['Sales']
        .sum()
        .sort_values(ascending=False)
        .head(5)
    )

    # Print result
    print(f"\nCity: {city} | Total Sales: {total_sales:.2f}")
    print(city_data.reset_index().to_string(index=False))
    print("-" * 50)
```

```python
import pandas as pd

# Ensure dates are in datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Step 1: Calculate delivery days
df['Delivery_Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Step 2: Define High-Value Products
# Option 1: Products above a certain sales threshold
threshold = df['Sales'].quantile(0.8)  # Top 20% sales
high_value_df = df[df['Sales'] >= threshold]

# Step 3: Calculate average delivery time for each Ship Mode
avg_delivery_by_mode = df.groupby('Ship Mode')['Delivery_Days'].mean()

# Step 4: Calculate average delivery time for high-value products in each Ship Mode
avg_delivery_high_value = high_value_df.groupby('Ship Mode')['Delivery_Days'].mean()

# Step 5: Compare them
comparison = pd.DataFrame({
    'Overall Avg Delivery (Days)': avg_delivery_by_mode,
    'High Value Avg Delivery (Days)': avg_delivery_high_value
}).fillna(0)

print(comparison)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df4 = pd.read_excel('/content/data for power bi.xlsx')  # Adjust the path if needed

# Clean and convert the quarter strings to proper Period format
df4['year and quarter'] = df4['year and quarter'].str.replace(' ', '')
df4['year and quarter'] = pd.PeriodIndex(df4['year and quarter'], freq='Q').astype(str)

# Calculate TOTAL SALES (sum of all sub-categories per quarter)
total_sales = df4.groupby('year and quarter')['Sales'].sum().reset_index()

# Plot
fig, ax = plt.subplots(figsize=(14, 7), facecolor='#e7e8f5')  # Outer background
ax.set_facecolor('#e7e8f5')  # Chart background

ax.plot(total_sales['year and quarter'], total_sales['Sales'],
        label='Total Sales', color='#414fb1', marker='o', linestyle='-', linewidth=2)

# Add top center text (without disturbing the line)
plt.text(0.1, 0.8,
         "Quarter 4 consistently shows highest sales\nwhile Quarter 1 has the lowest for each year.",
         ha='left', va='top',
         fontsize=14, weight='bold',
         fontfamily='Segoe UI',
         transform=ax.transAxes)

# Find min/max sales and the last quarter
max_sales = total_sales['Sales'].max()
last_quarter = total_sales['year and quarter'].iloc[-1]

# Annotate Q4 and Q1
for i, quarter in enumerate(total_sales['year and quarter']):
    sales_value = total_sales['Sales'].iloc[i]

    if quarter.endswith('Q4'):
        xytext = (-10, 5) if quarter == last_quarter and sales_value == max_sales else (0, 15)
        ax.annotate(f"Q4 Peak: ${sales_value:,.0f}", (quarter, sales_value),
                    textcoords="offset points", xytext=xytext,
                    ha='center', fontsize=9, color='green', weight='bold',
                    arrowprops=dict(arrowstyle='->', lw=0.5))

    elif quarter.endswith('Q1'):
        xytext = (5, -12) if i == 0 else (0, -15)
        ax.annotate(f"Q1 Low: ${sales_value:,.0f}", (quarter, sales_value),
                    textcoords="offset points", xytext=xytext,
                    ha='center', fontsize=9, color='red', weight='bold',
                    arrowprops=dict(arrowstyle='->', lw=0.5))

# ax.set_title('Total Sales Trend (Q4 Peaks vs Q1 Lows)')
ax.set_xlabel('Year and Quarter')
ax.set_ylabel('Total Sales ($)')
# Remove top and right borders (spines)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.tick_params(axis='x', rotation=45)
# ax.legend()
# ax.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
